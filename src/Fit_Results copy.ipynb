```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def plot_aggregated_nash_lottery(
    save_dir,
    n_clients_list,
    max_depths,
    approach='fedfor',
    title='Aggregated Nash & Lottery Occurrences'
):

    y_min = 0
    y_max = 101

    static_sums = []
    lottery_sums = []

    for n_clients in n_clients_list:
        s_total, l_total = 0, 0
        for max_depth in max_depths:
            fname = (
                f"Nash_Counts_{approach}"
                f"_nclients_{n_clients}"
                f"_maxiter_{max_depth}.csv"
            )
            df = pd.read_csv(os.path.join(save_dir, fname))
            s_total += int(df.loc[0, 'Static_Occurrences'])
            l_total += int(df.loc[0, 'Lottery_Occurrences'])
        static_sums.append(s_total)
        lottery_sums.append(l_total)

    plt.figure(figsize=(8, 6))
    plt.plot(
        n_clients_list, static_sums,
        linestyle='-', marker='s',
        color='brown',
        label='Static'
    )
    plt.plot(
        n_clients_list, lottery_sums,
        linestyle='--', marker='*',
        color='blue',
        label='Lottery'
    )

    plt.xlabel('Number of Clients')
    plt.ylabel('Total Occurrences')
    plt.title(title)
    plt.ylim(y_min, y_max)      
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
